```python
import os

image_dir = r'W:\Ellieshka\Documents\UCSD\MS in CSS\Degree Courses\4. Spring 26\COGS 202\cogs202_project\THINGS-database\osfstorage\object_images'

stimuli = []
for category in os.listdir(image_dir):
    category_path = os.path.join(image_dir, category)
    if os.path.isdir(category_path):
        for filename in os.listdir(category_path):
            if filename.endswith(('.jpg', '.png')):
                stimuli.append(os.path.join(category_path, filename))

print(len(stimuli))
```

```python
from collections import defaultdict

concept_images = defaultdict(list)

for path in stimuli:
    concept = os.path.basename(os.path.dirname(path))
    concept_images[concept].append(path)

print(len(concept_images))   # number of concepts
print(list(concept_images.keys())[:10])
```
